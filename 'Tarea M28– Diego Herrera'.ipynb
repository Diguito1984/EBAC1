```python
# Leer archivo JSON
try:
    with open(ruta_json, 'r', encoding='utf-8') as file:
        datos_json = json.load(file)
    print("‚úÖ Archivo JSON le√≠do exitosamente")
    
    # Determinar estructura del JSON
    if isinstance(datos_json, list):
        print(f"üìä Estructura: Lista con {len(datos_json)} elementos")
        if len(datos_json) > 0:
            print(f"üìã Keys del primer elemento: {list(datos_json[0].keys())}")
    else:
        print(f"üìä Estructura: Diccionario con keys: {list(datos_json.keys())}")
        
except Exception as e:
    print(f"‚ùå Error leyendo JSON: {e}")
    # Datos de ejemplo
    datos_json = [
        {'character': 'Tyrion', 'line': 'I drink and I know things.', 'season': 1},
        {'character': 'Cersei', 'line': 'When you play the game of thrones...', 'season': 1},
        {'character': 'Ned', 'line': 'Winter is coming.', 'season': 1}
    ] * 10
```

```python
# Convertir a DataFrame para mejor visualizaci√≥n
if isinstance(datos_json, list):
    df_json = pd.DataFrame(datos_json)
    print(f"üìä Dimensiones DataFrame: {df_json.shape[0]} filas x {df_json.shape[1]} columnas")
else:
    df_json = pd.DataFrame([datos_json])
    print("üìä JSON convertido a DataFrame de 1 fila")
```

```python
# Guardar en lista
if isinstance(datos_json, list):
    lista_json = datos_json
else:
    lista_json = [datos_json]
print(f"üìã Lista JSON creada con {len(lista_json)} elementos")
```

```python
# Primeras 10 filas
print("üîù PRIMERAS 10 FILAS JSON:")
if isinstance(datos_json, list) and len(datos_json) >= 10:
    primeras_10_json = pd.DataFrame(datos_json[:10])
    display(primeras_10_json)
elif isinstance(datos_json, list):
    primeras_10_json = pd.DataFrame(datos_json)
    display(primeras_10_json)
else:
    primeras_10_json = pd.DataFrame([datos_json])
    display(primeras_10_json)

# %%
# √öltimas 10 filas
print("üîö √öLTIMAS 10 FILAS JSON:")
if isinstance(datos_json, list) and len(datos_json) > 10:
    ultimas_10_json = pd.DataFrame(datos_json[-10:])
    display(ultimas_10_json)
else:
    print("‚ö†Ô∏è No hay suficientes datos para mostrar 10 filas finales")
    if isinstance(datos_json, list) and len(datos_json) > 0:
        display(pd.DataFrame(datos_json))
```

```python
# Seleccionar columnas y guardar nuevo archivo
if isinstance(datos_json, list) and len(datos_json) > 0:
    # Obtener todas las keys disponibles
    todas_keys = set()
    for item in datos_json:
        if isinstance(item, dict):
            todas_keys.update(item.keys())
    
    todas_keys = list(todas_keys)
    num_columnas = min(4, len(todas_keys))
    columnas_seleccionadas_json = todas_keys[:num_columnas]
```

```python
datos_reducidos = []
for item in datos_json:
    if isinstance(item, dict):
        nuevo_item = {key: item.get(key) for key in columnas_seleccionadas_json}
        datos_reducidos.append(nuevo_item)
```

```python
ruta_nuevo_json = os.path.join(ruta_base, "got_reducido.json")
with open(ruta_nuevo_json, 'w', encoding='utf-8') as file:
    json.dump(datos_reducidos, file, indent=2, ensure_ascii=False)

print(f"üíæ Nuevo archivo JSON guardado: '{ruta_nuevo_json}'")
print(f"üìã Columnas seleccionadas: {columnas_seleccionadas_json}")
```
