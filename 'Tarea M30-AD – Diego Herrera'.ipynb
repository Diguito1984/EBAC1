```python
print("Dimensiones del dataset:", df.shape)
print("\nColumnas disponibles:")
print(df.columns.tolist())
```

```python
# Mostrar primeras 20 filas
print("Primeras 20 filas del dataset:")
display(df.head(20))

# Mostrar últimas 5 filas
print("\nÚltimas 5 filas del dataset:")
display(df.tail(5))

# Mostrar sample aleatorio de 10 filas
print("\nMuestra aleatoria de 10 filas:")
display(df.sample(10, random_state=42))
```

```python
# Mostrar información de tipos de datos
print("Tipos de datos en el dataset:")
print(df.dtypes)

# Mostrar estadísticas descriptivas
print("\nEstadísticas descriptivas:")
display(df.describe())

# Verificar valores nulos
print("\nValores nulos por columna:")
null_counts = df.isnull().sum()
display(null_counts[null_counts > 0])
```

```python
# Identificar columnas que parecen numéricas pero no lo son
print("Columnas que podrían ser numéricas:")
potential_numeric_cols = []
for col in df.columns:
    if df[col].dtype == 'object':
        # Verificar si la columna contiene principalmente números
        sample_values = df[col].dropna().sample(min(10, len(df[col].dropna())), random_state=42)
        if all(sample_values.astype(str).str.replace('.', '', 1).str.replace(',', '', 1).str.replace('€', '', 1).str.replace('K', '', 1).str.replace('M', '', 1).str.strip().str.isdigit()):
            potential_numeric_cols.append(col)

print(potential_numeric_cols)

# Convertir columnas a numéricas
# Vamos a convertir 'Value', 'Wage' y 'Release Clause' que contienen valores monetarios
def convert_currency(value):
    if isinstance(value, str):
        value = value.replace('€', '')
        if 'M' in value:
            return float(value.replace('M', '')) * 1000000
        elif 'K' in value:
            return float(value.replace('K', '')) * 1000
        else:
            return float(value)
    return value

# Aplicar conversión a las columnas monetarias
df['Value'] = df['Value'].apply(convert_currency)
df['Wage'] = df['Wage'].apply(convert_currency)
df['Release Clause'] = df['Release Clause'].apply(convert_currency)

# Verificar que ahora son numéricas
print("\nTipos después de la conversión:")
print(f"Value: {df['Value'].dtype}")
print(f"Wage: {df['Wage'].dtype}")
print(f"Release Clause: {df['Release Clause'].dtype}")

# Mostrar estadísticas descriptivas con las nuevas columnas numéricas
print("\nEstadísticas descriptivas incluyendo columnas convertidas:")
display(df[['Value', 'Wage', 'Release Clause']].describe())
```

```python
# Verificar la columna Joined
print("Valores únicos en la columna 'Joined':")
print(df['Joined'].unique()[:10])

# Convertir la columna Joined a datetime
df['Joined'] = pd.to_datetime(df['Joined'], errors='coerce')

# Calcular años jugando (asumiendo que el año actual es 2023)
current_year = 2023
df['Years Playing'] = current_year - df['Joined'].dt.year

# Mostrar resultados
print("\nColumna 'Years Playing' añadida:")
display(df[['Name', 'Joined', 'Years Playing']].head(10))

# Estadísticas de la nueva columna
print("\nEstadísticas de 'Years Playing':")
print(df['Years Playing'].describe())
```

```python
# Buscar jugadores de México
mexican_players = df[df['Nationality'] == 'Mexico']

print(f"Jugadores de México encontrados: {len(mexican_players)}")
display(mexican_players[['Name', 'Age', 'Club', 'Overall', 'Value', 'Wage']])

# Mostrar algunos jugadores destacados
top_mexican_players = mexican_players.nlargest(5, 'Overall')
print("\nTop 5 jugadores mexicanos por valoración:")
display(top_mexican_players[['Name', 'Age', 'Club', 'Overall', 'Value', 'Wage']])
```

```python
# Ordenar por Release Clause (cláusula de rescisión)
sorted_by_release = df.sort_values('Release Clause', ascending=False)

print("Top 10 jugadores con mayor cláusula de rescisión:")
display(sorted_by_release[['Name', 'Nationality', 'Club', 'Overall', 'Value', 'Release Clause']].head(10))

print("\nTop 10 jugadores con menor cláusula de rescisión:")
display(sorted_by_release[['Name', 'Nationality', 'Club', 'Overall', 'Value', 'Release Clause']].tail(10))
```

```python
# Crear dataset agrupado por año de incorporación (Joined)
players_by_year = df.groupby(df['Joined'].dt.year).size().reset_index(name='Player Count')

# Renombrar columna
players_by_year = players_by_year.rename(columns={'Joined': 'Year'})

# Mostrar el nuevo dataset
print("Número de jugadores por año de incorporación:")
display(players_by_year)

# Estadísticas del nuevo dataset
print(f"\nAños cubiertos: {players_by_year['Year'].min()} - {players_by_year['Year'].max()}")
print(f"Total de años con datos: {len(players_by_year)}")
print(f"Promedio de jugadores por año: {players_by_year['Player Count'].mean():.2f}")
```
