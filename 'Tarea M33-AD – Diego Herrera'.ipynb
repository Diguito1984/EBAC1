```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
plt.style.use('default')
sns.set_palette("husl")

# Lectura del archivo
ruta_archivo = r'C:\Users\diego\OneDrive\Desktop\ebac\VisualCode\vgsales.csv'
df = pd.read_csv(ruta_archivo)

# 1) TOTAL DE VENTAS POR AÑO (desglosado por regiones)
print("1) TOTAL DE VENTAS POR AÑO (en millones)")
print("=" * 60)
ventas_por_anio = df.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].sum()
print(ventas_por_anio)
print("\n")

# 2) JUEGOS MÁS VENDIDOS POR AÑO (global)
print("2) JUEGOS MÁS VENDIDOS POR AÑO (Global)")
print("=" * 60)
idx = df.groupby('Year')['Global_Sales'].idxmax()
juegos_mas_vendidos = df.loc[idx, ['Year', 'Name', 'Platform', 'Genre', 'Global_Sales']]
print(juegos_mas_vendidos.set_index('Year'))
print("\n")

# 3) PLATAFORMAS CON MÁS INGRESOS POR AÑO
print("3) PLATAFORMAS CON MÁS INGRESOS POR AÑO (Global)")
print("=" * 60)
platform_ventas = df.groupby(['Year', 'Platform'])['Global_Sales'].sum().reset_index()
idx_platform = platform_ventas.groupby('Year')['Global_Sales'].idxmax()
top_platforms = platform_ventas.loc[idx_platform]
print(top_platforms.set_index('Year'))
print("\n")

# 4) GRÁFICO DE COMPARACIÓN DE VENTAS ENTRE GÉNEROS POR AÑO
plt.figure(figsize=(14, 8))
ventas_genero_anio = df.groupby(['Year', 'Genre'])['Global_Sales'].sum().unstack().fillna(0)
ventas_genero_anio.plot(kind='area', alpha=0.8, stacked=True)
plt.title('Evolución de Ventas Globales por Género de Videojuegos (1980-2020)')
plt.xlabel('Año')
plt.ylabel('Ventas Globales (millones)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# INSIGHTS: 2 GRÁFICOS ADICIONALES INTERESANTES

# Gráfico 1: Top 10 Plataformas con más ventas globales
plt.figure(figsize=(12, 6))
top_platforms = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)
top_platforms.plot(kind='barh', color='skyblue')
plt.title('Top 10 Plataformas con Más Ventas Globales')
plt.xlabel('Ventas Globales (millones)')
plt.ylabel('Plataforma')
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)
plt.show()

# Gráfico 2: Distribución de ventas por región a lo largo del tiempo
plt.figure(figsize=(12, 6))
regiones = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
ventas_regionales = df.groupby('Year')[regiones].sum()
ventas_regionales.plot(kind='line', linewidth=2.5)
plt.title('Evolución de Ventas por Región (1980-2020)')
plt.xlabel('Año')
plt.ylabel('Ventas (millones)')
plt.legend(['Norte América', 'Europa', 'Japón', 'Otros'])
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import pandas as pd
import numpy as np

# Lectura del archivo
ruta_archivo = r'C:\Users\diego\OneDrive\Desktop\ebac\VisualCode\vgsales.csv'
df = pd.read_csv(ruta_archivo)

# Filtrar últimos 4 años disponibles
ultimo_anio = df['Year'].max()
ultimos_4_anios = df[df['Year'] >= (ultimo_anio - 3)]

# 1) CRECIMIENTO POR GÉNERO
ventas_genero = ultimos_4_anios.groupby(['Year', 'Genre'])['Global_Sales'].sum().unstack().fillna(0)
crecimiento_genero = ((ventas_genero.loc[ultimo_anio] - ventas_genero.loc[ultimo_anio-3]) / ventas_genero.loc[ultimo_anio-3] * 100)
print("1) GÉNERO CON MAYOR Y MENOR CRECIMIENTO:")
print(f"   Mayor: {crecimiento_genero.idxmax()} ({crecimiento_genero.max():.1f}%)")
print(f"   Menor: {crecimiento_genero.idxmin()} ({crecimiento_genero.min():.1f}%)")
print()

# 2) CRECIMIENTO POR PLATAFORMA
ventas_platform = ultimos_4_anios.groupby(['Year', 'Platform'])['Global_Sales'].sum().unstack().fillna(0)
crecimiento_platform = ((ventas_platform.loc[ultimo_anio] - ventas_platform.loc[ultimo_anio-3]) / ventas_platform.loc[ultimo_anio-3] * 100)
print("2) PLATAFORMA CON MAYOR Y MENOR CRECIMIENTO:")
print(f"   Mayor: {crecimiento_platform.idxmax()} ({crecimiento_platform.max():.1f}%)")
print(f"   Menor: {crecimiento_platform.idxmin()} ({crecimiento_platform.min():.1f}%)")
print()

# 3) CRECIMIENTO POR VIDEOJUEGO
ventas_juego = ultimos_4_anios.groupby(['Year', 'Name'])['Global_Sales'].sum().unstack().fillna(0)
juegos_presentes = ventas_juego.columns[ventas_juego.loc[ultimo_anio-3].notna() & ventas_juego.loc[ultimo_anio].notna()]
crecimiento_juegos = ((ventas_juego.loc[ultimo_anio][juegos_presentes] - ventas_juego.loc[ultimo_anio-3][juegos_presentes]) / ventas_juego.loc[ultimo_anio-3][juegos_presentes] * 100)
print("3) VIDEOJUEGO CON MAYOR Y MENOR CRECIMIENTO:")
print(f"   Mayor: {crecimiento_juegos.idxmax()} ({crecimiento_juegos.max():.1f}%)")
print(f"   Menor: {crecimiento_juegos.idxmin()} ({crecimiento_juegos.min():.1f}%)")
print()

# 4) OPORTUNIDADES DE MERCADO
print("4) OPORTUNIDADES DE MERCADO:")
# Oportunidad 1: Géneros con alto crecimiento y baja saturación
participacion_genero = ultimos_4_anios.groupby('Genre')['Global_Sales'].sum()
generos_oportunos = crecimiento_genero[(crecimiento_genero > 30) & (participacion_genero < participacion_genero.median())]
print(f"   • Géneros emergentes: {', '.join(generos_oportunos.index[:3].tolist())}")

# Oportunidad 2: Regiones con crecimiento acelerado
ventas_region = ultimos_4_anios.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
crecimiento_region = ((ventas_region.loc[ultimo_anio] - ventas_region.loc[ultimo_anio-3]) / ventas_region.loc[ultimo_anio-3] * 100)
print(f"   • Región de mayor crecimiento: {crecimiento_region.idxmax().replace('_Sales', '')} ({crecimiento_region.max():.1f}%)")
```
