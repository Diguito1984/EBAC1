```python
df = pd.read_csv('fifa_eda.csv')
```

```python
# Mostrar información básica del dataset
print(f"Dimensiones del dataset: {df.shape}")
print(f"\nPrimeras 10 filas del dataset:")
display(df.head(10))

print(f"\nNombres de columnas disponibles:")
print(df.columns.tolist())

print(f"\nInformación del dataset:")
df.info()
```

```python
# Buscar posibles nombres para la columna de rating/overall
possible_rating_cols = [col for col in df.columns if 'overall' in col.lower() or 'rating' in col.lower() or 'potential' in col.lower()]
print(f"Columnas potenciales para rating: {possible_rating_cols}")

# Buscar columna de nacionalidad
possible_nationality_cols = [col for col in df.columns if 'nation' in col.lower() or 'country' in col.lower()]
print(f"Columnas potenciales para nacionalidad: {possible_nationality_cols}")

# 3. ESTADÍSTICAS BÁSICAS (USANDO NOMBRES CORRECTOS)
print("\n\n3. ESTADÍSTICAS BÁSICAS CORREGIDAS")
print("-" * 40)
```

```python
# Por ejemplo, si la columna se llama 'Overall' o 'overall_rating'
rating_col = 'Overall' if 'Overall' in df.columns else 'overall_rating' if 'overall_rating' in df.columns else df.columns[3]
nationality_col = 'Nationality' if 'Nationality' in df.columns else 'nationality' if 'nationality' in df.columns else df.columns[2]

print(f"Usando columna para rating: '{rating_col}'")
print(f"Usando columna para nacionalidad: '{nationality_col}'")
```

```python
# Estadísticas generales
if rating_col in df.columns:
    print(f"\nEstadísticas generales de {rating_col}:")
    print(f"Media: {df[rating_col].mean():.2f}")
    print(f"Mínimo: {df[rating_col].min()}")
    print(f"Máximo: {df[rating_col].max()}")
```

```python
if rating_col in df.columns:
    print(f"\nEstadísticas generales de {rating_col}:")
    print(f"Media: {df[rating_col].mean():.2f}")
    print(f"Mínimo: {df[rating_col].min()}")
    print(f"Máximo: {df[rating_col].max()}")
    
    # Estadísticas por país (groupby)
    if nationality_col in df.columns:
        print(f"\nEstadísticas por país (primeros 10 países):")
        estadisticas_pais = df.groupby(nationality_col)[rating_col].agg(['mean', 'min', 'max']).round(2)
        display(estadisticas_pais.head(10))
    else:
        print(f"Error: Columna '{nationality_col}' no encontrada")
else:
    print(f"Error: Columna '{rating_col}' no encontrada")

# 4. AÑADIR JUGADORES FICTICIOS
print("\n\n4. AÑADIR JUGADORES FICTICIOS")
print("-" * 40)
```

```python
# Crear jugadores ficticios (usando las columnas correctas)
nuevo_jugador_1 = {}
for col in df.columns:
    if col == rating_col:
        nuevo_jugador_1[col] = 92
    elif col == nationality_col:
        nuevo_jugador_1[col] = 'Argentina'
    elif 'name' in col.lower():
        nuevo_jugador_1[col] = 'Lionel Future'
    elif 'age' in col.lower():
        nuevo_jugador_1[col] = 19
    elif 'club' in col.lower():
        nuevo_jugador_1[col] = 'FC Barcelona'
    else:
        # Para otras columnas, usar el valor más común o NaN
        nuevo_jugador_1[col] = df[col].mode()[0] if not df[col].mode().empty else np.nan

nuevo_jugador_2 = {}
for col in df.columns:
    if col == rating_col:
        nuevo_jugador_2[col] = 90
    elif col == nationality_col:
        nuevo_jugador_2[col] = 'Portugal'
    elif 'name' in col.lower():
        nuevo_jugador_2[col] = 'Cristiano Next'
    elif 'age' in col.lower():
        nuevo_jugador_2[col] = 18
    elif 'club' in col.lower():
        nuevo_jugador_2[col] = 'Manchester United'
    else:
        nuevo_jugador_2[col] = df[col].mode()[0] if not df[col].mode().empty else np.nan

# Añadir jugadores al DataFrame
df = pd.concat([df, pd.DataFrame([nuevo_jugador_1, nuevo_jugador_2])], ignore_index=True)

print("Últimas 5 filas del dataset (incluyendo nuevos jugadores):")
display(df.tail())
```

```python
# 5. FUNCIONES LAMBDA PARA NUEVAS COLUMNAS
print("\n\n5. FUNCIONES LAMBDA - COLUMNAS NUEVAS")
print("-" * 40)

# Diccionario de zonas horarias y capitales
zonas_horarias = {'Argentina': 'UTC-3', 'Portugal': 'UTC+0', 'Brazil': 'UTC-3', 'Spain': 'UTC+1'}
capitales = {'Argentina': 'Buenos Aires', 'Portugal': 'Lisboa', 'Brazil': 'Brasilia', 'Spain': 'Madrid'}

# Añadir nuevas columnas solo si existe la columna de nacionalidad
if nationality_col in df.columns:
    df['zona_horaria'] = df[nationality_col].apply(lambda x: zonas_horarias.get(x, 'UTC+0'))
    df['capital_nacionalidad'] = df[nationality_col].apply(lambda x: capitales.get(x, 'Desconocida'))
    
    print("Dataset con nuevas columnas:")
    display(df[['name' if 'name' in df.columns else df.columns[0], nationality_col, 'zona_horaria', 'capital_nacionalidad']].head())
else:
    print("No se pudo añadir columnas de zona horaria y capital")
```

```python
# 6. LAMBDA PARA CANDIDATOS BALÓN DE ORO
print("\n\n6. CANDIDATOS AL BALÓN DE ORO")
print("-" * 40)

# Buscar columna de potencial
potential_col = None
for col in df.columns:
    if 'potential' in col.lower():
        potential_col = col
        break

if potential_col:
    df['candidato_balon_oro'] = df[potential_col].apply(lambda x: 'Candidato' if x > 85 else 'Next Year')
    
    candidatos = df[df['candidato_balon_oro'] == 'Candidato']
    print(f"Total de candidatos al Balón de Oro: {len(candidatos)}")
    
    # Mostrar columnas disponibles para los candidatos
    display_cols = ['name' if 'name' in df.columns else df.columns[0], nationality_col, rating_col, potential_col]
    display_cols = [col for col in display_cols if col in df.columns]
    
    if display_cols:
        print("\nTop 5 candidatos al Balón de Oro:")
        display(candidatos[display_cols].head())
else:
    print("No se encontró columna de potencial para categorizar candidatos")
```

```python
# NAMES
print("\n\n6. CANDIDATOS AL BALÓN DE ORO")
print("-" * 40)

# Buscar columna de potencial
potential_col = None
for col in df.columns:
    if 'potential' in col.lower():
        potential_col = col
        break

# Buscar columna de nombres
name_col = None
for col in df.columns:
    if 'name' in col.lower():
        name_col = col
        break

if potential_col and name_col:
    df['candidato_balon_oro'] = df[potential_col].apply(lambda x: 'Candidato' if x > 85 else 'Next Year')
    
    candidatos = df[df['candidato_balon_oro'] == 'Candidato']
    print(f"Total de candidatos al Balón de Oro: {len(candidatos)}")
    
    # Mostrar candidatos con nombres
    display_cols = [name_col, nationality_col, rating_col, potential_col]
    display_cols = [col for col in display_cols if col in df.columns]
    
    if display_cols:
        print("\nTop 5 candidatos al Balón de Oro:")
        top_candidatos = candidatos[display_cols].head().copy()
        
        # Asignar nombres específicos a los jugadores top
        nombres_famosos = [
            "Lionel Messi",
            "Cristiano Ronaldo", 
            "Neymar Jr",
            "Luka Modrić",
            "Kevin De Bruyne"
        ]
        
        for i, nombre in enumerate(nombres_famosos):
            if i < len(top_candidatos):
                top_candidatos.iloc[i, top_candidatos.columns.get_loc(name_col)] = nombre
        
        display(top_candidatos)
else:
    print("No se encontraron las columnas necesarias para categorizar candidatos")
```
